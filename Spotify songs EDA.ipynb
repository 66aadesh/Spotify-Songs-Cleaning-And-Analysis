# Handling Missing Values

```python
# Print the missing values in all columns
missing_values = spotify_df.isnull().sum()
missing_values
```

```python
# Find the 1.5 times of average no. of missing value
(sum(missing_values)/len(list(missing_values)))*1.5
```

```python
# Finding the columns with missing values more than the average no. of missing value
new_dict = dict(missing_values)

for key,value in new_dict.items():
    if value > 1179:
        print(key)
```

```python
# Dropping the above columns, cleaning the df

print('All Columns=',len(spotify_df.columns))

spotify_df = spotify_df.drop(['SiriusXM Spins','Pandora Track Stations','Soundcloud Streams','TIDAL Popularity'], axis=1)

print('Cleaned Columns=',len(spotify_df.columns))
```

```python
# Further Cleaning
# The columns with relatively less known platform will be dropped.
# ISRC column is a fact column and doesnt represent the trends or any measure of popularity or merit

spotify_df = spotify_df.drop(['Apple Music Playlist Count', 'AirPlay Spins','ISRC','Deezer Playlist Count','Deezer Playlist Reach', 'Amazon Playlist Count', 'Pandora Streams','Shazam Counts'], axis=1)

print('Cleaned Columns=',len(spotify_df.columns))
```

```python
# Change datatype appropriately
# Convert 'Release Date' to datetime
spotify_data_df['Release Date'] = pd.to_datetime(spotify_data_df['Release Date'])

# Change datetime format to 'dd-mm-yyyy'
spotify_data_df['Release Date'] = spotify_data_df['Release Date'].dt.strftime('%d-%m-%y')

# Change datatypes of other columns
# Remove commas from the columns and convert to the appropriate datatype
spotify_data_df['All Time Rank'] = spotify_data_df['All Time Rank'].str.replace(',', '').astype(int)
spotify_data_df['Spotify Streams'] = spotify_data_df['Spotify Streams'].str.replace(',', '').astype(float)
spotify_data_df['Spotify Playlist Count'] = spotify_data_df['Spotify Playlist Count'].str.replace(',', '').astype(float)
spotify_data_df['Spotify Playlist Reach'] = spotify_data_df['Spotify Playlist Reach'].str.replace(',', '').astype(float)

# Adjust pandas display settings to avoid scientific notation
pd.set_option('display.float_format', '{:f}'.format)

spotify_data_df.dtypes
```
